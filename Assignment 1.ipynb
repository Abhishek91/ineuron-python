Que 2: Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in acomma-separated sequence on a single line.

```python
#Option 1
numList=[]
for num in range(2000, 3200):
    if (num%7==0) and (num%5!=0):
        numList.append(str(num))
        
print(','.join(numList))
```

```python
#Option 2
data = [str(num) for num in range(2000, 3200) if num%7==0 and num%5!=0]
print(','.join(data))
```

Que: 4 Write a Python program to find the volume of a sphere with diameter 12 cm.
Formula: V=4/3 * π * r 3

```python
PI=3.14
diameter = 12
radius = diameter//2
v=4/3 * PI * radius ** 3
print(v)
```
